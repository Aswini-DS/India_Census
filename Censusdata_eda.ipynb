```python
censusdata.replace('-','0',inplace=True)
```

```python
censusdata['Urban'] = censusdata['Urban'].fillna('0')
censusdata['Urban'] = censusdata['Urban'].astype(int)
censusdata['Rural'] = censusdata['Rural'].fillna('0')
censusdata['Rural'] = censusdata['Rural'].astype(int)
```

```python
#As Total Population is divided into Urban and Rural, Filling Urban NaN values into Population-Rural and vise-versa for Urban

censusdata['Urban'].replace(0,censusdata.Persons - censusdata.Rural,inplace=True)
```

```python
censusdata['Rural'].replace(0,censusdata.Persons - censusdata.Urban,inplace=True)
```

## __Education Level in India__

```python
censusdata['Graduate.and.Above'].fillna(0,inplace=True)
censusdata['Below.Primary'].fillna(0,inplace=True)
censusdata['Primary'].fillna(0,inplace=True)
censusdata['Middle'].fillna(0,inplace=True)
censusdata['Matric.Higher.Secondary.Diploma'].fillna(0,inplace=True)
```

```python
labels = 'Graduate.and.Above','Below.Primary', 'Primary', 'Middle','Matric.Higher.Secondary.Diploma'
explode = (0.1, 0, 0, 0,0)
education = [sum(censusdata['Graduate.and.Above']),sum(censusdata['Below.Primary']),sum(censusdata['Primary']),sum(censusdata['Middle']),sum(censusdata['Matric.Higher.Secondary.Diploma'])]
plt.pie(education,explode=explode, labels=labels,autopct='%1.1f%%')
```

```python
censusdata.groupby('State')[["Graduate.and.Above","Below.Primary","Primary","Middle","Matric.Higher.Secondary.Diploma"]].sum().plot(kind='bar',stacked=True,figsize=(15,10))
```

#### As per above plot Daman and Diu , Dadra and Nagar Haveli and Lakshdweep are having very low education facilities

```python
literate = [sum(censusdata['Males..Literate']),sum(censusdata['Females..Literate'])]
labels =  'Male Literate','Female Literate'
plt.pie(literate,labels=labels,autopct='%1.1f%%')
```

```python
censusdata.groupby('State').apply(lambda x: x['Persons..literacy.rate'].sum() / x['Persons'].sum()).plot(kind='bar',figsize=(10,10))
```

#### As per above plots literacy rate per person is very low in West Bangal , Bihar and Andhra

```python
censusdata.groupby('State').apply(lambda x: x['Drinking.water.facilities'].sum() / x['Persons'].sum()).plot(kind='bar',figsize=(10,10))
```

#### As per above plots Delhi and Chandigarh are having low Driniking water facilities

```python
censusdata.groupby('State')[["Permanent.House","Semi.permanent.House","Temporary.House"]].sum().plot(kind='bar',stacked=True,figsize=(15,10))
```

```python
censusdata.groupby('State').apply(lambda x: x['Post..telegraph.and.telephone.facility'].sum() / x['Persons'].sum()).plot(kind='bar',figsize=(10,10))
```

#### As per above plots HP is having better communication facilities

```python
censusdata.groupby('State')[["Sex.ratio..females.per.1000.males."]].sum().plot(kind='bar',figsize=(10,10))
```
